***Few shot promting***

```python
# Define the price per unit for each customer type
price_domestic = 2 # rupees per unit
price_commercial = 3 # rupees per unit
price_industrial = 5 # rupees per unit

# Get the number of units consumed from the user
units_consumed_input = input("Enter the number of units consumed: ")
units_consumed = int(units_consumed_input)

# Get the customer type from the user (1 for Domestic, 2 for Commercial, 3 for Industrial)
customer_type_input = input("Enter customer type (1: Domestic, 2: Commercial, 3: Industrial): ")
customer_type = int(customer_type_input)

# Calculate and print the price based on customer type
print("\n--- Energy Pricing Details ---")
print(f"Units Consumed: {units_consumed}")
print("----------------------------")

if customer_type == 1:
    price_to_pay = units_consumed * price_domestic
    print(f"Domestic Customer Price: {price_to_pay} Rupees")
elif customer_type == 2:
    price_to_pay = units_consumed * price_commercial
    print(f"Commercial Customer Price: {price_to_pay} Rupees")
elif customer_type == 3:
    price_to_pay = units_consumed * price_industrial
    print(f"Industrial Customer Price: {price_to_pay} Rupees")
else:
    print("Invalid customer type entered. Please choose 1, 2, or 3.")

print("----------------------------")
```
